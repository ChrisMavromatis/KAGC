```python
def LateMat (df, e_cutoff):
    x = df.FCounts()[(df.FCounts().date3 > e_cutoff) | (df.FCounts().date4 > e_cutoff)]
    return x
```

```python
def GenPlot(df, save_plot, out_ti,  fext, start, stop):
    """
    Parameters:
        * df = dataframe of DU material class
        * TITLE = text plot title
        * out_ti = figure title
        * save _plot = save the plot 'Y' or 'N'
        * fext = output file extension (eg '.png')
        * start = earliest date in data set (round it to the nearest century)
        * stop latest date in the dataset (round it to nearest century)
    Requires: seaborn, matplotlib.pyplot and numpy
    """
    import  matplotlib.pyplot as plt
    import matplotlib.patches as mpatches
    import seaborn as sns
    import numpy as np


    df['lbl'] = df.DU +'-'+ df.Cat.fillna(df.mcls)

    x2 = df

    my_range = range(1, len(x2.index)+1)

    x_ticks = []
    for i in range(start, stop, 100):
        x_ticks.append(i)

    # Plot style
    if len(x2) > 15:
        plt.figure(figsize=(10,6))
    else:
        pass

    if len(x2) > 20:
        plt.yticks(my_range, x2.lbl, fontsize=12)
    else:
       
        plt.tick_params(axis = 'y', rotation = 0, labelsize = 11)


    sns.set_style("white", rc= {"xtick.bottom": True, "ytick.left": True,})
    sns.set_context("paper", font_scale = 0.9)
    plt.hlines(y=my_range, xmin=x2['date3'], xmax=x2['date4'], color='black', alpha= 1)
    plt.scatter(x2['date3'], my_range, color='black', alpha= 1, label='date3', s = 11)


    # PLot lables
    plt.yticks(my_range, x2.lbl, fontsize=9)
    plt.xticks(x_ticks, fontsize = 10)
    plt.tick_params(axis = 'x', rotation = 0)
    plt.tick_params(axis = 'y', rotation = 0, labelsize = 9)
    #plt.title("Times New Roman 18", fontname = "Times New Roman", fontsize=18)
    plt.xlabel('Production Spans', fontsize=12)
    plt.ylabel('DU Material Class', fontsize=12)
    
    # Plot historical events as blocks
    plt.axvspan(350, 400, alpha =0.2, color= "red")
    earthquake = mpatches.Patch(color= "red", alpha=0.2, label = " Beginning of the \n Early Byzantine \n Tectonic Paroxysm")

    plt.axvspan(430, 450, alpha = 0.2, color = "green") 
    recon =  mpatches.Patch(color= "green", alpha=0.2, label = " Beginning of Episcopal Precinct \n and House of Eustolios's \n reconstruction?")

    abandon = plt.axvspan(685, 700, alpha = 0.2, color = "blue", label = " Beginning of \n Kourion's abandoment")

    nero =  plt.axvspan(54, 117, alpha = 0.2, color = "yellow", label = " The Reigns of Nero-Trajan")

    severians = plt.axvspan(145, 211, alpha = 0.2, color = "gray", label = " The Severan Dynasty")
    plt.legend(loc ="lower center", bbox_to_anchor= (0.5, -0.3), ncol = 4, handles = [nero, severians, earthquake, recon, abandon], fontsize =10)

    #save options
    if save_plot == 'Y':
         plt.tight_layout()
         plt.savefig(out_ti + fext, dpi = 1200)
    else:
        pass
```

```python
def FC_DatePlt_B(df, Vlines, save_plot, out_ti,  fext, start, stop):
    """
    Plots the date spans of of the FCounts method from the DU class

    Parameters:
        * df = = DU class object
        * Vlines = add mean and TPQ lines 'Y' or 'N'
        * out_ti = figure title
        * save _plot = save the plot 'Y' or 'N'
        * fext = output file extension (eg '.png')
    Requires: seaborn, matplotlib.pyplot and numpy
    """
    import  matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np


    x = df.FCounts().sort_values(by=['date3', 'date4'])

    my_range = range(1, len(x.index)+1)

    x_ticks = []
    for i in range(start, stop, 100):
        x_ticks.append(i)

    # Plot style
    sns.set_style("white", rc= {"xtick.bottom": True, "ytick.left": True,})
    sns.set_context("paper", font_scale = 0.9)

    if len(x) > 20:
        plt.figure(figsize=(10,6))
    else:
        pass


    plt.hlines(y=my_range, xmin=x['date3'], xmax=x['date4'], color='black', alpha= 1)
    plt.scatter(x['date3'], my_range, color='black', alpha= 1, label='date3', s = 11)

    # Plot lables
    if len(x) > 20:
        plt.yticks(my_range, x.Seq, fontsize=12)
    else:
        plt.yticks(my_range, x.mcls +"-"+ x.Seq.astype(str), fontsize=12)
        plt.tick_params(axis = 'y', rotation = 0, labelsize = 11)
        

    

    plt.xticks(x_ticks, fontsize=10)
    plt.tick_params(axis = 'x', rotation = 0)
    plt.tick_params(axis = 'y', rotation = 0, labelsize = 8)
    #plt.title("Times New Roman 18", fontname = "Times New Roman", fontsize=18)
    plt.xlabel('Production Spans', fontsize=12)
    plt.ylabel('Type-Form', fontsize=12)
    plt.tight_layout()

    # Additinal 

    if Vlines == 'Y':
        # Mean Median
        plt.vlines(x = df.cdu.MPDate.mean(), ymin= 0, ymax = len(x.diag), color='blue', alpha= 1, linestyles=['--'])
        #plt.text(df.cdu.MPDate.mean() + 1, len(x.diag)+1, "MD", rotation= 0, color = 'blue')

        # TPQ
        plt.vlines(x = x.date3.max(), ymin= 0, ymax = len(x.diag), color='red', alpha= 1, linestyles=['--'])
        #plt.text(x.date3.max() + 1, len(x.diag)+1, "TPQ", rotation= 0, color ='red')
    else:
        pass

    
    # Save
    if save_plot == 'Y':
         plt.savefig(out_ti + fext, dpi = 1200)
    else:
        pass

```

#### <a id='toc3_2_3_1_'></a>[DUs A009 and 603 compared](#toc0_)

A review of the latest items by material class for each DU indicates relatively good inter-DU agreement between their latest glass, coins, and fine ware. However, there is no inter-DU agreement between their lamps and amphorae. In DU A009, the latest amphora’s opening production date is in the early second century A.D.; in DU A603, it is the mid-fifth century A.D. DU A009’s latest lamp has an opening production date in the mid-first century A.D.; in contrast, DU A603’s latest lamp has its opening production date in the early fourth century A.D. 
Items from Marker Groups 1 and 3 (MG1 and MG3; see Chapter 3.3) constitute the most significant parts of DU A009 and DU A603’s assemblages. DU A603’s assemblage is among the largest, richest, and most diverse in the study area. Its assemblage has a pronounced MG1 component represented by items associated with the architectural, display, production, tools, and special purpose functional groups. In DU A009, most MG1 items belong to the architectural, display, and production functional groups. MG3 items include a range of storage/processing, transport/storage, table ware, and illumination items; transport/storage (i.e., amphorae) is the richest of these components in both DUs. However, with only 13 unique amphora types, DU A009’s amphora assemblage is not as diverse or chronologically expansive as the 25 unique forms in DU A603. These DUs share nine amphorae forms which date mainly to the Hellenistic to Mid-Roman periods. After pooling the assemblages, we find that the DU A603 assemblage contains over 85% (86.2%, n= 25) of the pooled forms while DU A007 contains c. 45% (44.8%, n = 13). Apart from its typological range, DU A603’s amphora assemblage has only two forms, LRA 4 and LRA 1, with opening production dates in the fourth or fifth centuries A.D. These forms are not present in DU A009. MG3 also includes a substantial quantity of animal bone (969 fragments from DU A603 and 124 fragments from DU A009), some of which preserved evidence of burning, butchering, and bone working (Volume 2, Chapter 4). 

In contrast, MG2 items are not well-represented in DU A009 or DU A603. They are limited to a few personal adornments (e.g., rings), personal items (e.g., coins), and special purpose items (e.g., terracotta figurine fragments and an incense burner fragment). Like DUs A020 and A019, the low frequency of personal items and adornments is consistent with what we might expect from items incorporated into these deposits through casual loss in a domestic context. Both DUs have a residual component consisting predominantly of Hellenistic and early Roman items with production dates earlier than 24 B.C. In DU A603, the ratio of non-residual to non-residual items is c. 3:1; in DU A009, it is c. 6:1. Both values are comparable to DU A019 and DU A020’s in neighboring TrA05.

While Depositional Units A009 and A603 follow the same broad trend, comparing components of their assemblage profiles, especially the amphorae, has given us some idea of how the cross-cutting in DU A603 likely conflated several dumping episodes, including the one that produced DU A009. Based on DU A009 and A603’s aggregate chronological and assemblage characteristics, we can broadly interpret them as admixtures of earthquake clean-up debris and post-quake municipal refuse. The residual items suggest that material derived from Hellenistic and Early Roman deposits, most likely from the Acropolis, also contributed to the waste streams that generated DUs A603 and A009. The comparatively high quantities of marble reduction waste, animal bone (including butchered and worked fragments), and transport/storage vessels indicate that domestic and commercial waste streams also played an essential role in shaping this sector of Area A’s stratigraphic biography. Like DUs A019 and A020, Marker Group 2 items are not well represented in DUs A009 and A603. Carbon buildup, sharp breaks, and in-context joins from DU A009 indicate that this material did not remain in the systemic context long before redeposition. Nevertheless, at some point after c. 450 A.D., DUs A009 and A603 entirely obscured the remaining parts of FA07 that had remained above ground to this point. Based on the available stratigraphic sections and photographs, it is clear that these deposits continue further to TrA01 and TrA01-EE’s south and east.
