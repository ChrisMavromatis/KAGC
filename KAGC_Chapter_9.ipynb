```python
def GenPlot(df, save_plot, out_ti,  fext, start, stop):
    """
    Parameters:
        * df = dataframe of DU material class
        * TITLE = text plot title
        * out_ti = figure title
        * save _plot = save the plot 'Y' or 'N'
        * fext = output file extension (eg '.png')
        * start = earliest date in data set (round it to the nearest century)
        * stop latest date in the dataset (round it to nearest century)
    Requires: seaborn, matplotlib.pyplot and numpy
    """
    import  matplotlib.pyplot as plt
    import matplotlib.patches as mpatches
    import seaborn as sns
    import numpy as np


    df['lbl'] = df.DU +'-'+ df.Cat.fillna(df.mcls)

    x2 = df

    my_range = range(1, len(x2.index)+1)

    x_ticks = []
    for i in range(start, stop, 100):
        x_ticks.append(i)

    # Plot style
    if len(x2) > 15:
        plt.figure(figsize=(10,6))
    else:
        pass

    if len(x2) > 20:
        plt.yticks(my_range, x2.lbl, fontsize=12)
    else:
       
        plt.tick_params(axis = 'y', rotation = 0, labelsize = 11)


    sns.set_style("white", rc= {"xtick.bottom": True, "ytick.left": True,})
    sns.set_context("paper", font_scale = 0.9)
    plt.hlines(y=my_range, xmin=x2['date3'], xmax=x2['date4'], color='black', alpha= 1)
    plt.scatter(x2['date3'], my_range, color='black', alpha= 1, label='date3', s = 11)


    # PLot lables
    plt.yticks(my_range, x2.lbl, fontname = "Sans Serif", fontsize=9)
    plt.xticks(x_ticks, fontsize = 10)
    plt.tick_params(axis = 'x', rotation = 0)
    plt.tick_params(axis = 'y', rotation = 0, labelsize = 9)
    #plt.title("Times New Roman 18", fontname = "Times New Roman", fontsize=18)
    plt.xlabel('Production Spans', fontname = "Sans Serif", fontsize=12)
    plt.ylabel('DU Material Class', fontname = "Sans Serif", fontsize=12)
    
    # Plot historical events as blocks
    plt.axvspan(350, 400, alpha =0.2, color= "red")
    earthquake = mpatches.Patch(color= "red", alpha=0.2, label = " Beginning of the \n Early Byzantine \n Tectonic Paroxysm")

    plt.axvspan(430, 450, alpha = 0.2, color = "green") 
    recon =  mpatches.Patch(color= "green", alpha=0.2, label = " Beginning of Episcopal Precinct \n and House of Eustolios's \n reconstruction?")

    abandon = plt.axvspan(685, 700, alpha = 0.2, color = "blue", label = " Beginning of \n Kourion's abandoment")

    nero =  plt.axvspan(54, 117, alpha = 0.2, color = "yellow", label = " The Reigns of Nero-Trajan")

    severians = plt.axvspan(145, 211, alpha = 0.2, color = "gray", label = " The Severan Dynasty")
    plt.legend(loc ="lower center", bbox_to_anchor= (0.5, -0.3), ncol = 4, handles = [nero, severians, earthquake, recon, abandon], fontsize =10)

    #save options
    if save_plot == 'Y':
         plt.tight_layout()
         plt.savefig(out_ti + fext, dpi = 600)
    else:
        pass
```

# <a id='toc4_'></a>[The TA01 Area: TA01, TrA02, TrA07, and TrA08](#toc0_)
KAGC personnel identified TA01 while excavating TrA02 during the 1998 field season. The discovery of
TA01, and a blank tomb (FA10) immediately to its south, suggested that the area around TrA02 had high
potential for containing additional cist tombs. As a result, during the 1999 field season, KAGC team
members expanded their investigations around TrA02’s south and east, with TrA08, and to its west, with
TrA07 (Fig. 9.1; Fig. 9.9). Cumulatively, these trenches cover an area of about 60 m2, or approximately
25% of Area A’s total excavated area. Whi|le these excavations did not uncover additional cist tombs, they
did identify several contexts from these trenches that are groupable into three discrete DUs: A014, A015,
A017 (Table 9.1). Perhaps the most interesting deposit from around TA01 is the dark stain feature, FA04,
which KAGC team members recorded during their excavation of TrA08’s Context 004. For the present
purpose, we may note that over 200 glass shards, with some deformed by prolonged heat exposure,
were recovered from the comparatively small area that FA04 occupied in TrA08. We may also note that
the DUs surrounding TA01 differed in their sediment and assemblage compositions from DU A001 from
TrA03 and TrA05 as well as DUs A002, A003, and A013 which are associated with the kiln complex
(Vol.1's Chapter 10.3) further to the east.

## <a id='toc4_1_'></a>[DUS A501](#toc0_)
 
 KAGC personnel identified one context (Context 005 in Trench TrA02’s stratigraphic sequence) within TA01. As the only layer from within TA01, it is also the only Contributing Context (CC) associated with Depositional Unit A501 (DU A501). DU A501’s sediment consisted of c. 0.40–0.45 m of very loose yellowish-brown to brownish-gray sandy silt. Although DU A501’s sediment was similar to DU A801’s, it lacked the clast and irregular limestone spall inclusions that characterize the latter. The excavators also noted extensive root intrusions and a moderate number of snail shells in the tomb’s northern and western sectors. DU A501 abutted all of Tomb TA01’s interior walls and directly overlaid its floor. The DU’s interfaces with these interior surfaces were all sharply defined. Before excavating within Tomb TA01, KAGC team members also noted that numerous human bones were resting on and protruding through DU A501’s upper interface. However, most of the human bone was concentrated within c. 0.20 m of the tomb’s floor. The human remains consisted of c. 501 bone and bone fragments clustered in and around the tomb’s northwestern sector. These fragments are attributable to an MNI (Minimum Number of Individuals) of four, including one adult male (23–57 years), two females (27–31 and 15–45 years) (Volume 2, Chapter 2, Table 2.5) and one indeterminate individual.

## <a id='toc4_3_'></a>[DU A014](#toc0_)

DU A014 consists of three Contributing Contexts (CC): Trench TrA02’s Context 001, Trench TrA07’s Context 001, and Trench TrA08’s Context 001. DU A014’s sediment consisted of a grayish-brown to light-brown silty sand with graded compaction. The sediment contained numerous comparatively small angular limestone inclusions and numerous rootlets but few artifacts. DU A014 completely overlaid DU A015 in Trenches TrA07 and TrA08. However, in Trench TrA02’s eastern sector, DU A014 was exceptionally shallow (< 0.10 m) and it overlaid worked bedrock, an incomplete cist tomb (FA10), and DU A015. DU A014 also abutted against the bedrock spur wall. The interface between DU A014 and the spur wall and underlying bedrock, including FA10, was sharply defined; the interface between DUs A014 and A015 was clearly defined (transitional zone: 2.5–6.0 cm).

trenches (i.e. 5 x 5 m) (see Fig. 9.15 for a map, Fig. 5.9 for a photograph)

Here we may also note that nearly all the sediment in the trenches from Area

underlying bedrock surface; these deposits are all Contributing Contexts (CCs) to DU B901 (Chapter 10.5).
While the bulldozer truncated most of the sediment in Area B’s eastern quadrant, several trenches were
free from modern material; these deposits are all CCs to DU B002 (Fig. 9.21).

Area B’s most important component is the 21 cist tombs the project excavated between 1995 and
1998. Based on an analysis of their centroids, well over half of the tombs fall within one standard
deviation from the group’s mean center (Fig. 9.15; Table 9.3). The cist tombs are, on average, located
2.50 m from their nearest neighbor. This value, which is slightly larger than the expected random mean

## <a id='toc5_1_'></a>[BDS-1](#toc0_)

Area B’s Depositional Sequence 1 (BDS-1) combines DU B002, trench layers without modern artifacts, and DU B501, the upper layer in nearly every cist tomb. Its DU B002 and DU B501 components provide 10 and 20 Contributing Contexts, respectively. The sediment from both components consisted primarily of a semi-compact grayish-brown sandy silt. While angular and rounded rubble and tomb lid fragments were less common in DU B002, they were a significant inclusion type in DU B501 (e.g., TB18). In some cases (e.g., Tomb TB01), tomb lid fragments extended into the underlying sediment (i.e., DUs B502 and B503). Because the construction of the theater road in the late 1980s had truncated BDS-1’s DU B002 component, its sediment was relatively shallow (Chapter 3). The interface between BDS-1’s DU B002 component and the underlying worked bedrock was sharply defined. It also partially or entirely overlaid several features in Trenches TrB03–08 and Trench TrB13. These include blank tombs and libation pipe sockets. In Trench TrB09 (below), BDS-1 overlaid that trench’s Context 003 and abutted against FB19, a mud-bonded wall segment, and their interfaces were clearly defined. BDS-1’s DU B501 component varied in thickness (i.e., 0.05 - 020 m). In six cist tombs (e.g., TB03), BDS-1’s DU B501 component was the only identifiable layer, filling the tombs completely. Its interface with the tomb floors was sharply defined. In the remaining cases, it either overlaid DU B502 or DU B503; its interface with these DUs was either clearly or sharply defined.
